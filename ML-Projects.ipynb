```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('Copy of sonar data.csv', header=None)
```

```python
data
```

```python
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
data[60].value_counts()
```

```python
data.groupby(60).mean()
```

```python
x = data.drop(columns=60, axis=1)
y = data[60]
```

```python
print(x)
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.1,random_state=1,stratify=y)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
x_train_pred = lr.predict(x_train)
traing_data_accuracy = accuracy_score(x_train_pred,y_train)
```

```python
print('Accuracy_score_on_training data :', traing_data_accuracy)
```

```python
x_test_pred = lr.predict(x_test)
test_data_accuracy = accuracy_score(x_test_pred,y_test)
```

```python
print('Accuracy_score_on_test data :', test_data_accuracy)
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
diabetes = pd.read_csv('diabetes.csv')
```

```python
diabetes.head()
```

```python
diabetes.shape
 
```

```python
diabetes.describe()
```

```python
diabetes['Outcome'].value_counts()
```

```python
diabetes.groupby('Outcome').mean()
```

```python
x = diabetes.drop(columns='Outcome', axis=1)
y = diabetes['Outcome']
```

```python
print(x)
```

```python
print(y)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
standardized_data = scaler.transform(x)
```

```python
x= standardized_data
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, stratify=y, random_state=2)
```

```python
x.shape,x_train.shape, x_test.shape
```

```python
svc = svm.SVC(kernel='linear')
```

```python
svc.fit(x_train,y_train)
```

```python
x_train_predic = svc.predict(x_train)
training_data_accuracy = accuracy_score(x_train_predic, y_train)
```

```python
print('accuracy_score of training data:', training_data_accuracy)
```

```python
x_test_predic = svc.predict(x_test)
testing_data_accu = accuracy_score(x_test_predic,y_test)
```

```python
print('accuracy_score of training data:', testing_data_accu)
```

```python
input_data = (3,158,76,36,245,31.6,0.851,28)
inputdataas_numarry = np.asarray(input_data)
inputdata_reshaped = inputdataas_numarry.reshape(1,-1)
std_data = scaler.transform(inputdata_reshaped)
print(std_data)

prediction = svc.predict(std_data)
print(prediction)

if prediction[0]==0:
    print('The person is not diabetic')
else:
    print('The person is diabetic')
```
